```python
basic_pathes = [] # list of dirs containing output dirs named problem-model-hp.trial.
token = False
do_over = True
interest = ['directory', "model", "hp", "trial", "steps", "file", "shard", "seq_acc", "ch_acc"]
sort_order = ['directory', 'model', 'hp', 'trial', 'steps', 'shard']
def show_df(df):
    df_a = df.sort_values(sort_order)    
    for terms, df_m in df_a.groupby(["model", "hp"]):
        model, hp = terms
        display(df_m[interest])
def get_summary_statistics(df):
    summary_statistics = ["model", "hp", "shard", "steps", "mean_seq", "std_seq", "max_seq", "#trials"]
    summary_df = []
    for cat, df_m in df.groupby(["model", "hp", "shard", "steps"]):
        mean_seq = df_m['seq_acc'].mean()
        std_seq = df_m['seq_acc'].std()    
        max_seq = df_m['seq_acc'].max()
        trials = len(df_m)
        summary_df.append(list(cat) + [mean_seq, std_seq, max_seq, trials])
    summary_df = pd.DataFrame(summary_df, columns=summary_statistics)
    return summary_df        
```

```python
problem = "number_sequence_prediction_third"
model = "seq_in_grid_out_architecture"  # A list of models or a model
hp = None  # A list of hps or a hp
df = gather_all_statistics_by_problem(basic_pathes, problem, model, hp, token=token, do_over=do_over)
get_summary_statistics(df)
```
